```python
import os
from getpass import getpass
import openai

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY") or getpass(
    "Enter OpenAI API Key: "
)
```

```python
openai_model = "gpt-4o-mini"
```

```python
from langchain_openai import ChatOpenAI

# For normal accurate responses
llm = ChatOpenAI(temperature=0.0, model=openai_model, openai_api_key = os.environ["OPENAI_API_KEY"])
```

```python
from typing_extensions import TypedDict

class State(TypedDict):
    graph_state: str
    graph_ticker: str
```

```python
def node_1(state):
    print("---Node 1---")
    return {"graph_state": state['graph_state']}

def node_2(state):
    print("---Analyst Valuation---")
    ticker = yf.Ticker(state['graph_ticker'])
    df = ticker.analyst_price_targets
    current = df['current']
    return {"graph_state": current}

def node_3(state):
    print("---Sustainability---")
    ticker = yf.Ticker(state['graph_ticker'])
    df = ticker.sustainability
    
    response = llm(f"""using this data frame {df} tell me if this company is a ethically 
                    sustainable company, I would also like you to format it as 
                    ?/10 for topic, this is due to...""")

    return {"graph_state": response.content}

def node_4(state):
    print("---Profitability---")
    ticker = yf.Ticker(state['graph_ticker'])
    df = ticker.financials
    
    response = llm(f"""using this data frame {df} tell me if this company is a 
                    profitable company, I would also like you to format it as 
                    ?/10 for topic, this is due to...""")

    return {"graph_state": response.content}
```

```python
from typing import Literal

def decide_mood(state) -> Literal["node_2", "node_3", "node_4"]:
    
    # Often, we will use state to decide on the next node to visit
    user_input = state['graph_state'] 
    print(user_input)
    # Here, let's just do a 50 / 50 split between nodes 2, 3
    if user_input == "profit":

        # 50% of the time, we return Node 2
        return "node_4"
    
    if user_input == "price":

        # 50% of the time, we return Node 4
        return "node_2"
    
    if user_input == "sustain":
    # 50% of the time, we return Node 3
        return "node_3"
```

```python
from IPython.display import Image, display
from langgraph.graph import StateGraph, START, END
```

```python
# Build graph
builder = StateGraph(State)

builder.add_node("node_1", node_1)
builder.add_node("node_2", node_2)
builder.add_node("node_3", node_3)
builder.add_node("node_4", node_4)

# Logic
builder.add_edge(START, "node_1")
builder.add_conditional_edges("node_1", decide_mood)
builder.add_edge("node_2", END)
builder.add_edge("node_3", END)
builder.add_edge("node_4", END)

# Add
graph = builder.compile()

# View
display(Image(graph.get_graph().draw_mermaid_png()))
```

## Super Agent

```python
from typing_extensions import TypedDict

class State(TypedDict):
    response: str
    final_answer: str
    question: str
    ticker: str
    next_action: str
    history: str
```

```python
def assistant_node(state):
    print("---   Assistant Node   ---")
    tools = """"Stock_Price_Value_Node", "Sustainability_Node", "Financials_Node", "Finalise_Node", "Cashflow_Node"""""
    response = llm(f"""Based on the question, do you believe this can be answered using 
                   the current content, 
                   if this is the case, please reply
                   with "Finalise_Node", otherwise, think based on what tools we have available
                   could help you find the correct answer, and answer as the name of the 
                   tool, here is the list of tools {tools}, here is the question 
                   {state['question']} here is the response so far 
                   {state['final_answer']}, if you have used all the tools in the
                   history, enter the "Finalise_Node" here is the history {state['history']}""")
    print(response.content)
    return {"response": response.content, "history": state['history'] + "    " + response.content}
```

```python
def Valuation(state):
    print("---Analyst Valuation---")
    ticker = yf.Ticker(state['ticker'])
    df = ticker.analyst_price_targets
    answer = df.values
    response = llm(f"""This is the table of contents: {df}, Summarise the table of contents, make sure we do not get rid of any details that might help with answering questions on this text.""")
    return {"final_answer": df, "history": state['history'] + response.content}

def Sustainability(state):
    print("---Sustainability---")
    ticker = yf.Ticker(state['ticker'])
    df = ticker.sustainability
    response = llm(f"""This is the table of contents: {df}, Summarise the table of contents, make sure we do not get rid of any details that might help with answering questions on this text.""")

    answer = df.values

    return {"final_answer": df, "history": state['history'] + response.content}

def Financials(state):
    print("---Financials---")
    ticker = yf.Ticker(state['ticker'])
    df = ticker.financials
    answer = df.values
    response = llm(f"""This is the table of contents: {df}, Summarise the table of contents, make sure we do not get rid of any details that might help with answering questions on this text.""")

    return {"final_answer": df, "history": state['history'] + response.content}

def Cashflow(state):
    print("---Cashflow---")
    ticker = yf.Ticker(state['ticker'])
    df = ticker.cash_flow

    answer = df.values
    response = llm(f"""This is the table of contents: {df}, Summarise the table of contents, make sure we do not get rid of any details that might help with answering questions on this text.""")

    return {"final_answer": df, "history": state['history'] + response.content}

def Finalise(state):
    print("---Final---")
    print(state['final_answer'])
    response = llm(f"""using the current context {state['final_answer']} try to 
                   answer the question {state['question']}, do not lie.""")
    print(state['history'])

    return {"final_answer": response, "history": state['history'] + response.content}
```

```python
from typing import Literal

def decide_node(state) -> Literal["Stock_Price_Value_Node", "Sustainability_Node", "Financials_Node", "Finalise_Node", "Cashflow_Node"]:
    print("Deciding")
    # Often, we will use state to decide on the next node to visit
    response_input = state['response'] 
    print(response_input)
    # Here, let's just do a 50 / 50 split between nodes 2, 3
    return response_input
```

```python
from IPython.display import Image, display
from langgraph.graph import StateGraph, START, END
```

```python
# Build graph
builder = StateGraph(State)

builder.add_node("Assistant_Node", assistant_node)
builder.add_node("Stock_Price_Value_Node", Valuation)
builder.add_node("Sustainability_Node", Sustainability)
builder.add_node("Financials_Node", Financials)
builder.add_node("Finalise_Node", Finalise)
builder.add_node("Cashflow_Node", Cashflow)

# Logic
builder.add_edge(START, "Assistant_Node")
builder.add_conditional_edges("Assistant_Node", decide_node)
builder.add_edge("Cashflow_Node", "Assistant_Node")
builder.add_edge("Stock_Price_Value_Node", "Assistant_Node")
builder.add_edge("Sustainability_Node", "Assistant_Node")
builder.add_edge("Financials_Node", "Assistant_Node")
builder.add_edge("Finalise_Node", END)

# Add
graph = builder.compile()

# View
display(Image(graph.get_graph().draw_mermaid_png()))
```
